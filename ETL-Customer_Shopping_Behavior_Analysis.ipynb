```python
#To check null values of all cols
df.isnull().sum()
```

```python
#From upper o/p-Review Rating has 37 Null values, so replace it with mean/median, mostly med is selected as mean could get affected by extreme vals & med is robust to outliers,
#Here we will select median - but not overall median of the whole col but its not the best practice, bcoz ratings can vary a lot as clothes & footwear will have different review ratings
#We will IMPUTE missing vals using median review rating within each category & replace that with missing value)
#Eg. If clothing item is missing Review Rating then it will be replaced by median Review Rating of missing clothing item not footwear item)
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
#This data is now cleaned in the right way
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#Feature Engineering- Add 1 col - age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group']=pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create column purchase_freq_days

frequency_mapping={
    'Fortnightly':14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-weekly': 14,
    'Annually': 365,
    'Every 3 months': 90
}

df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
    
```
